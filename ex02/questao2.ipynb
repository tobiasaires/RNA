## Questão 2

- Problema associado ao dataset escolhido;
`O dataset escolhido faz referência ao dataset MNIST, porém de uma maneira mais simplificada e reduzida. No original elas são representas 28x28, neste toy_dataset ele é somente 8x8`

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as matplot
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

datasets = load_wine()

X = datasets.data
y = datasets.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
classifier_1 = MLPClassifier(hidden_layer_sizes=10, learning_rate_init=0.3, max_iter=10)

classifier_1.fit(X_train, y_train)
```

```python
classifier_2 = MLPClassifier(hidden_layer_sizes=100, learning_rate_init=0.03, max_iter=100)

classifier_2.fit(X_train, y_train)
```

```python
classifier_3 = MLPClassifier(hidden_layer_sizes=1000, learning_rate_init=0.003, max_iter=1000)

classifier_3.fit(X_train, y_train)
```

- Plotar a evolução da função custo (loss) ao longo do treinamento (épocas) para verificar a
corretude do algoritmo de treinamento.

```python
loss_values_datasets_1 = classifier_1.loss_curve_
matplot.plot(loss_values_datasets_1)
matplot.show()
```

```python
loss_values_datasets_2 = classifier_2.loss_curve_
matplot.plot(loss_values_datasets_2)
matplot.show()
```

```python
loss_values_datasets_3 = classifier_3.loss_curve_
matplot.plot(loss_values_datasets_3)
matplot.show()
```

- Reportar taxas de acerto (ou erro) nos conjuntos de treinamento e teste para diferentes
configurações da rede neural.

```python
predict_1 = classifier_1.predict(X_test)
accuracy_score_1 = accuracy_score(y_test, predict_1)
print('accuracy_score_1 = {}'.format(accuracy_score_1))
```

```python
predict_2 = classifier_2.predict(X_test)
accuracy_score_2 = accuracy_score(y_test, predict_2)
print('accuracy_score_2 = {}'.format(accuracy_score_2))
```

```python
predict_3 = classifier_3.predict(X_test)
accuracy_score_3 = accuracy_score(y_test, predict_3)
print('accuracy_score_3 = {}'.format(accuracy_score_3))
```
